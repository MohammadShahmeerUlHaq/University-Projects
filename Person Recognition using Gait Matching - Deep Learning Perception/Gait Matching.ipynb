```python
import os
import numpy as np
import matplotlib.pyplot as plt
import random
import cv2
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, optimizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_fscore_support
from sklearn.manifold import TSNE
from tensorflow.keras.utils import to_categorical
import seaborn as sns
from tqdm import tqdm
from glob import glob
import warnings
import pickle
from collections import defaultdict

# Suppress warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
random_seed = 42
np.random.seed(random_seed)
tf.random.set_seed(random_seed)
random.seed(random_seed)

# Enable mixed precision training for better memory usage
from tensorflow.keras import mixed_precision
policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)

# For consistent GPU usage if available
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

class GaitRecognitionSystem:
    def __init__(self, data_path, img_height=64, img_width=44, num_known=20, num_unknown=5):
        """
        Initialize the Gait Recognition System with smaller default image sizes
        """
        self.data_path = data_path
        self.img_height = img_height
        self.img_width = img_width
        self.num_known = num_known
        self.num_unknown = num_unknown
        self.total_people = num_known + num_unknown
        self.model = None
        self.unknown_threshold = 0.7
        self.feature_extractor = None
        self.known_features = None

    def load_and_preprocess_data(self):
        print("Loading and preprocessing data...")

        # Get all person IDs (sorted)
        all_person_ids = sorted(os.listdir(self.data_path))[:self.total_people]

        # Get all image paths with labels (similar to your PyTorch code)
        data = []  # (img_path, label)
        for label_idx, person_id in enumerate(all_person_ids):
            person_dir = os.path.join(self.data_path, person_id)
            for cond_folder in os.listdir(person_dir):
                cond_path = os.path.join(person_dir, cond_folder)
                if os.path.isdir(cond_path):
                    for seq_folder in os.listdir(cond_path):
                        seq_path = os.path.join(cond_path, seq_folder)
                        if os.path.isdir(seq_path):
                            for img_file in os.listdir(seq_path):
                                if img_file.endswith('.png'):
                                    img_path = os.path.join(seq_path, img_file)
                                    data.append((img_path, label_idx))

        # Shuffle the data
        random.shuffle(data)

        # Split into known and unknown subjects (80-20 split per person)
        known_data = []
        unknown_data = []

        # Group by person for stratified splitting
        person_data = {i: [] for i in range(self.total_people)}
        for path, label in data:
            person_data[label].append((path, label))

        # Split each person's data 80-20
        for label in person_data:
            person_images = person_data[label]
            split_idx = int(0.8 * len(person_images))
            if label < self.num_known:  # Known subjects
                known_data.extend(person_images[:split_idx])
            else:  # Unknown subjects
                unknown_data.extend(person_images[:split_idx])

        # Process known data into sequences (16 frames per sequence)
        X_known = []
        y_known = []

        # Group by person and sequence to create temporal sequences
        person_seq_data = defaultdict(list)
        for img_path, label in known_data:
            # Extract sequence identifier from path
            seq_id = os.path.dirname(img_path)
            person_seq_data[(label, seq_id)].append(img_path)

        # Create sequences of 16 frames for known subjects
        for (label, _), img_paths in person_seq_data.items():
            img_paths = sorted(img_paths)  # Sort by frame number
            sequences = [img_paths[i:i+16] for i in range(0, len(img_paths), 16)]

            for seq in sequences:
                if len(seq) == 16:
                    processed_frames = []
                    for img_path in seq:
                        try:
                            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                            if img is None:
                                continue
                            img = cv2.resize(img, (self.img_width, self.img_height))
                            img = img / 255.0
                            processed_frames.append(img)
                        except Exception as e:
                            print(f"Error processing frame {img_path}: {e}")

                    if len(processed_frames) == 16:
                        X_known.append(np.stack(processed_frames))
                        y_known.append(label)

        # Process unknown data similarly
        X_unknown = []

        unknown_seq_data = defaultdict(list)
        for img_path, label in unknown_data:
            seq_id = os.path.dirname(img_path)
            unknown_seq_data[seq_id].append(img_path)

        for seq_id, img_paths in unknown_seq_data.items():
            img_paths = sorted(img_paths)
            sequences = [img_paths[i:i+16] for i in range(0, len(img_paths), 16)]

            for seq in sequences:
                if len(seq) == 16:
                    processed_frames = []
                    for img_path in seq:
                        try:
                            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                            if img is None:
                                continue
                            img = cv2.resize(img, (self.img_width, self.img_height))
                            img = img / 255.0
                            processed_frames.append(img)
                        except Exception as e:
                            print(f"Error processing frame {img_path}: {e}")

                    if len(processed_frames) == 16:
                        X_unknown.append(np.stack(processed_frames))

        # Convert to numpy arrays
        X_known = np.array(X_known)
        y_known = np.array(y_known)
        X_unknown = np.array(X_unknown)

        # Split known data into train and test
        X_train, X_test_known, y_train, y_test_known = train_test_split(
            X_known, y_known, test_size=0.2, random_state=42, stratify=y_known
        )

        # Create test set with known and unknown samples
        X_test = np.concatenate([X_test_known, X_unknown])
        y_test = np.concatenate([
            y_test_known,
            np.full(len(X_unknown), self.num_known)  # Label for unknown samples
        ])

        # Convert labels to categorical
        y_train = to_categorical(y_train, num_classes=self.num_known)
        y_test = to_categorical(y_test, num_classes=self.num_known + 1)

        print(f"Final dataset shapes:")
        print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
        print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")

        return X_train, y_train, X_test, y_test

    def sample_frames(self, frames, sequence_path, num_frames=16):
        """Same as before"""
        if len(frames) < 5:
            return None

        if len(frames) <= num_frames:
            selected_frames = frames
        else:
            step = len(frames) / num_frames
            indices = [int(i * step) for i in range(num_frames)]
            selected_frames = [frames[i] for i in indices]

        processed_frames = []
        for frame in selected_frames:
            img_path = os.path.join(sequence_path, frame)
            try:
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                if img is None:
                    continue
                img = cv2.resize(img, (self.img_width, self.img_height))
                img = img / 255.0
                processed_frames.append(img)
            except Exception as e:
                print(f"Error processing frame {img_path}: {e}")

        if len(processed_frames) < num_frames:
            if not processed_frames:
                return None
            last_frame = processed_frames[-1]
            processed_frames.extend([last_frame] * (num_frames - len(processed_frames)))

        return np.stack(processed_frames)


    def build_model(self):
        """Fixed model building with proper input specification"""
        input_layer = layers.Input(shape=(16, self.img_height, self.img_width))
        x = layers.Reshape((16, self.img_height, self.img_width, 1))(input_layer)

        # Conv Block 1
        x = layers.TimeDistributed(layers.Conv2D(16, (3,3), activation='relu', padding='same'))(x)
        x = layers.TimeDistributed(layers.BatchNormalization())(x)
        x = layers.TimeDistributed(layers.MaxPooling2D((2,2)))(x)

        # Conv Block 2
        x = layers.TimeDistributed(layers.Conv2D(32, (3,3), activation='relu', padding='same'))(x)
        x = layers.TimeDistributed(layers.BatchNormalization())(x)
        x = layers.TimeDistributed(layers.MaxPooling2D((2,2)))(x)

        # Conv Block 3
        x = layers.TimeDistributed(layers.Conv2D(64, (3,3), activation='relu', padding='same'))(x)
        x = layers.TimeDistributed(layers.BatchNormalization())(x)
        x = layers.TimeDistributed(layers.MaxPooling2D((2,2)))(x)

        # Flatten
        x = layers.TimeDistributed(layers.Flatten())(x)

        # LSTM layers
        x = layers.LSTM(128, return_sequences=True)(x)
        x = layers.LSTM(64)(x)

        # Dense layers
        x = layers.Dense(64, activation='relu')(x)
        x = layers.Dropout(0.5)(x)

        # Output
        output = layers.Dense(self.num_known, activation='softmax', dtype='float32')(x)

        # Create model
        model = models.Model(inputs=input_layer, outputs=output)

        model.compile(
            optimizer=optimizers.Adam(learning_rate=0.0005),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        model.summary()
        self.model = model
        return model

    def train_model(self, X_train, y_train, X_val, y_val, epochs=30, batch_size=16):
        """Training with smaller batch size"""
        print("Training model...")

        callbacks = [
            EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6),
            ModelCheckpoint('gait_model_best.h5', monitor='val_accuracy', save_best_only=True, mode='max')
        ]

        val_known_indices = np.argmax(y_val, axis=1) < self.num_known
        X_val_known = X_val[val_known_indices]
        y_val_known = y_val[val_known_indices]
        y_val_known = y_val_known[:, :self.num_known]

        history = self.model.fit(
            X_train, y_train,
            validation_data=(X_val_known, y_val_known),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=callbacks,
            verbose=1
        )

        return history

    def extract_features(self, X):
        """Fixed feature extraction with proper initialization"""
        if self.feature_extractor is None:
            # Warm up the model first
            if len(X) > 0:
                _ = self.model.predict(X[:1])

            # Create feature extractor
            self.feature_extractor = models.Model(
                inputs=self.model.input,
                outputs=self.model.layers[-2].output
            )

        return self.feature_extractor.predict(X)

    def calculate_unknown_threshold(self, X_train, y_train, X_val=None, y_val=None):
        """Same as before"""
        train_features = self.extract_features(X_train)
        self.known_features = train_features

        distances = []
        max_samples = 1000
        if len(train_features) > max_samples:
            indices = np.random.choice(len(train_features), max_samples, replace=False)
            sampled_features = train_features[indices]
        else:
            sampled_features = train_features

        for i in range(len(sampled_features)):
            for j in range(i+1, len(sampled_features)):
                dist = np.linalg.norm(sampled_features[i] - sampled_features[j])
                distances.append(dist)

        mean_dist = np.mean(distances)
        std_dist = np.std(distances)
        threshold = mean_dist + 2 * std_dist

        if X_val is not None and y_val is not None:
            val_features = self.extract_features(X_val)
            val_true_labels = np.argmax(y_val, axis=1) < self.num_known

            val_min_distances = []
            for i in range(len(val_features)):
                dists = [np.linalg.norm(val_features[i] - train_feature) for train_feature in sampled_features]
                val_min_distances.append(min(dists))

            thresholds = np.linspace(np.min(val_min_distances), np.max(val_min_distances), 100)
            best_f1 = 0
            best_threshold = threshold

            for t in thresholds:
                predicted_unknown = (np.array(val_min_distances) > t).astype(int)
                true_unknown = (~val_true_labels).astype(int)

                precision, recall, f1, _ = precision_recall_fscore_support(
                    true_unknown, predicted_unknown, average='binary', zero_division=0
                )

                if f1 > best_f1:
                    best_f1 = f1
                    best_threshold = t

            threshold = best_threshold

        self.unknown_threshold = threshold
        print(f"Unknown detection threshold set to: {threshold}")
        return threshold

    def predict_with_unknown_detection(self, X):
        """Same as before"""
        predictions = self.model.predict(X)
        features = self.extract_features(X)

        min_distances = []
        for feature in features:
            dists = [np.linalg.norm(feature - known_feature) for known_feature in self.known_features]
            min_distances.append(min(dists))

        final_predictions = []

        for i in range(len(predictions)):
            if min_distances[i] > self.unknown_threshold:
                pred = np.zeros(self.num_known + 1)
                pred[self.num_known] = 1.0
            else:
                confidence = np.max(predictions[i])
                pred = np.zeros(self.num_known + 1)

                if confidence > 0.5:
                    pred[:self.num_known] = predictions[i]
                else:
                    pred[self.num_known] = 1.0

            final_predictions.append(pred)

        return np.array(final_predictions)

    def evaluate_model(self, X_test, y_test):
        """Same as before"""
        print("Evaluating model...")

        y_pred = self.predict_with_unknown_detection(X_test)
        y_true = np.argmax(y_test, axis=1)
        y_pred_classes = np.argmax(y_pred, axis=1)

        class_names = [f"Person_{i+1}" for i in range(self.num_known)] + ["Unknown"]
        print("\nClassification Report:")
        print(classification_report(y_true, y_pred_classes, target_names=class_names))

        cm = confusion_matrix(y_true, y_pred_classes)

        plt.figure(figsize=(12, 10))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.title('Confusion Matrix')
        plt.tight_layout()
        plt.savefig('confusion_matrix.png')
        plt.show()

        known_true = y_true < self.num_known
        known_pred = y_pred_classes < self.num_known

        print("\nKnown vs Unknown Detection:")
        print(classification_report(known_true, known_pred, target_names=["Unknown", "Known"]))

        self.visualize_features(X_test, y_true)

        return {
            'accuracy': np.mean(y_true == y_pred_classes),
            'confusion_matrix': cm,
            'classification_report': classification_report(y_true, y_pred_classes, target_names=class_names, output_dict=True)
        }

    def visualize_features(self, X, y_true):
        """Same as before"""
        features = self.extract_features(X)
        tsne = TSNE(n_components=2, random_state=random_seed)
        features_2d = tsne.fit_transform(features)

        plt.figure(figsize=(12, 10))

        for i in range(self.num_known):
            idx = y_true == i
            plt.scatter(
                features_2d[idx, 0],
                features_2d[idx, 1],
                label=f"Person_{i+1}",
                alpha=0.7
            )

        idx = y_true == self.num_known
        plt.scatter(
            features_2d[idx, 0],
            features_2d[idx, 1],
            label="Unknown",
            marker='x',
            s=100,
            alpha=0.7
        )

        plt.title('t-SNE Visualization of Gait Features')
        plt.xlabel('t-SNE Component 1')
        plt.ylabel('t-SNE Component 2')
        plt.legend()
        plt.savefig('feature_visualization.png')
        plt.show()

    def save_model(self, filepath='gait_recognition_model'):
        """Same as before"""
        self.model.save(f'{filepath}.h5')

        with open(f'{filepath}_params.pkl', 'wb') as f:
            pickle.dump({
                'unknown_threshold': self.unknown_threshold,
                'img_height': self.img_height,
                'img_width': self.img_width,
                'num_known': self.num_known
            }, f)

        if self.known_features is not None:
            np.save(f'{filepath}_known_features.npy', self.known_features)

        print(f"Model saved to {filepath}")

    def load_model(self, filepath='gait_recognition_model'):
        """Same as before"""
        self.model = models.load_model(f'{filepath}.h5')

        self.feature_extractor = models.Model(
            inputs=self.model.input,
            outputs=self.model.layers[-2].output
        )

        with open(f'{filepath}_params.pkl', 'rb') as f:
            params = pickle.load(f)
            self.unknown_threshold = params['unknown_threshold']
            self.img_height = params['img_height']
            self.img_width = params['img_width']
            self.num_known = params['num_known']

        try:
            self.known_features = np.load(f'{filepath}_known_features.npy')
        except:
            print("Known features file not found.")

        print(f"Model loaded from {filepath}")

def main():
    print("Starting Gait Recognition System with Unknown Detection")

    # Initialize with smaller image sizes
    gait_system = GaitRecognitionSystem(
        data_path='/root/.cache/kagglehub/datasets/trnquanghuyn/casia-b/versions/1/output',
        img_height=64,  # Reduced from 128
        img_width=44,   # Reduced from 88
        num_known=20,
        num_unknown=5
    )

    X_train, y_train, X_test, y_test = gait_system.load_and_preprocess_data()

    # Build the optimized model
    gait_system.build_model()

    # Split test data
    X_val, X_test_final, y_val, y_test_final = train_test_split(
        X_test, y_test, test_size=0.5, random_state=42
    )

    # Train with smaller batch size
    history = gait_system.train_model(
        X_train, y_train,
        X_val, y_val,
        epochs=30,  # Reduced epochs for demo
        batch_size=16  # Reduced batch size
    )

    # Calculate threshold
    gait_system.calculate_unknown_threshold(X_train, y_train, X_val, y_val)

    # Evaluate
    metrics = gait_system.evaluate_model(X_test_final, y_test_final)
    print(f"Overall accuracy: {metrics['accuracy']:.4f}")

    # Save model
    gait_system.save_model()

    # Plot training history
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

    print("Training and evaluation complete!")

if __name__ == "__main__":
    main()
```
